```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from ipywidgets import interact, widgets, Layout
from IPython.display import display, clear_output

# --- 1. Generación de Datos Sintéticos ---
def generar_datos():
    np.random.seed(42)
    n_muestras = 500
    patogenos = ['A. baumannii', 'K. pneumoniae', 'S. aureus', 'E. coli']
    antibioticos = ['Meropenem', 'Ciprofloxacina', 'Amikacina', 'Tigeciclina']
    sedes = ['Norte', 'Sur', 'Centro', 'Occidente']

    data = {
        'ID_Muestra': range(1, n_muestras + 1),
        'Patógeno': np.random.choice(patogenos, n_muestras),
        'Antibiótico': np.random.choice(antibioticos, n_muestras),
        'Sede': np.random.choice(sedes, n_muestras),
        'Halo_Inhibicion': np.random.randint(6, 35, n_muestras), # en mm
        'Resistente': np.random.choice([True, False], n_muestras, p=[0.4, 0.6])
    }
    return pd.DataFrame(data)

df = generar_datos()

# --- 2. Widgets de Control ---
dropdown_patogeno = widgets.Dropdown(
    options=['Todos'] + list(df['Patógeno'].unique()),
    value='Todos',
    description='Patógeno:',
    style={'description_width': 'initial'}
)

dropdown_sede = widgets.Dropdown(
    options=['Todas'] + list(df['Sede'].unique()),
    value='Todas',
    description='Sede Clínica:',
    style={'description_width': 'initial'}
)

output = widgets.Output()

# --- 3. Función de Actualización Dinámica ---
def actualizar_dashboard(change):
    with output:
        clear_output(wait=True)

        # Filtrado lógico
        df_filt = df.copy()
        if dropdown_patogeno.value != 'Todos':
            df_filt = df_filt[df_filt['Patógeno'] == dropdown_patogeno.value]
        if dropdown_sede.value != 'Todas':
            df_filt = df_filt[df_filt['Sede'] == dropdown_sede.value]

        # Resumen de métricas
        total = len(df_filt)
        resistencia_media = (df_filt['Resistente'].mean() * 100) if total > 0 else 0

        print(f"📊 Total de Aislamientos: {total}  |  🧬 Resistencia Promedio: {resistencia_media:.1f}%")

        # Gráfico 1: Distribución de Halos (Histograma)
        fig1 = px.histogram(
            df_filt, x='Halo_Inhibicion', color='Antibiótico',
            nbins=20, barmode='overlay',
            title=f"Distribución de Halos de Inhibición - {dropdown_patogeno.value}",
            labels={'Halo_Inhibicion': 'Diámetro del Halo (mm)'},
            template='plotly_dark'
        )

        # Gráfico 2: Resistencia por Sede (Barras)
        df_res = df_filt.groupby('Sede')['Resistente'].mean().reset_index()
        df_res['Resistente'] *= 100
        fig2 = px.bar(
            df_res, x='Sede', y='Resistente',
            title="Porcentaje de Resistencia por Sede",
            labels={'Resistente': '% Resistencia'},
            color='Sede',
            template='plotly_dark'
        )

        # Mostrar gráficos
        fig1.show()
        fig2.show()

# Enlazar los widgets a la función
dropdown_patogeno.observe(actualizar_dashboard, names='value')
dropdown_sede.observe(actualizar_dashboard, names='value')

# --- 4. Mostrar el Layout ---
controles = widgets.HBox([dropdown_patogeno, dropdown_sede])
dashboard = widgets.VBox([controles, output])

display(dashboard)
actualizar_dashboard(None) # Ejecución inicial
```
